# Condiciones de frontera

**Coincidencia de valor**  
_para 0<S<L*:_
$$
\begin{aligned}
V(L^{\ast})
&= K-\frac{2r}{2r+\sigma^{2}}\,K \\
&= K\!\left(1-\frac{2r}{2r+\sigma^{2}}\right) \\
&= K\!\left(\frac{2r+\sigma^{2}-2r}{2r+\sigma^{2}}\right) \\
&= \boxed{\frac{K\,\sigma^{2}}{\,2r+\sigma^{2}\,}} \, .
\end{aligned}
$$

_para S>L*:_
$$
\begin{aligned}
V(S)
&= \frac{K\sigma^{2}}{2r+\sigma^{2}}
\left(\frac{2r+\sigma^{2}}{2rK}\,S\right)^{-\frac{2r}{\sigma^{2}}} \\
V(L^{\ast})
&= \frac{K\sigma^{2}}{2r+\sigma^{2}}
\left(\frac{2r+\sigma^{2}}{2rK}\,\frac{2r}{2r+\sigma^{2}}\,K\right)^{-\frac{2r}{\sigma^{2}}} \\
&= \frac{K\sigma^{2}}{2r+\sigma^{2}}\,(1)^{-\frac{2r}{\sigma^{2}}}
= \boxed{\frac{K\,\sigma^{2}}{\,2r+\sigma^{2}\,}} \, .
\end{aligned}
$$


**Smooth pasting**

Para 0 < S ≤ L* simplemente se deriva y se obtiene claramente -1, para el caso S > L* desarrollamos:


$$
\begin{aligned}
V'(L^{\ast})
&= \frac{K\sigma^{2}}{2r+\sigma^{2}}\,
\left(-\frac{2r}{\sigma^{2}}\right)\,
\left(\frac{2r+\sigma^{2}}{2rK}\right)\,
\left(\frac{2r+\sigma^{2}}{2rK}\,L^{\ast}\right)^{-\frac{2r}{\sigma^{2}}-1} \\[6pt]
&= (-1)\,
\frac{K\sigma^{2}}{2r+\sigma^{2}}\,
\frac{2r}{\sigma^{2}}\,
\frac{2r+\sigma^{2}}{2rK}\,
\left(\frac{2r+\sigma^{2}}{2rK}\cdot
\frac{2r}{2r+\sigma^{2}}\,K\right)^{-\frac{2r}{\sigma^{2}}-1} \\[6pt]
&= (-1)\,
\underbrace{\frac{2rK\sigma^{2}}{2rK\sigma^{2}}}_{=\,1}\,
\underbrace{\frac{2r+\sigma^{2}}{2r+\sigma^{2}}}_{=\,1}\,
\underbrace{1^{-\frac{2r}{\sigma^{2}}-1}}_{=\,1}
\;=\; -1 \, .
\end{aligned}
$$

**Límite en infinito**

Como $s\to\infty$, usamos la rama $s\ge L^{\ast}$:

$$
\lim_{s\to\infty}
\frac{K\sigma^{2}}{2r+\sigma^{2}}
\left(\frac{2r+\sigma^{2}}{2r}\,\frac{s}{K}\right)^{-\frac{2r}{\sigma^{2}}}
=0.
$$

Al tender $s$ a infinito, el término entre paréntesis $\to\infty$ y, como el exponente es negativo, la potencia $\to 0$. Al multiplicar por la constante positiva $\tfrac{K\sigma^{2}}{2r+\sigma^{2}}$, el límite total es $0$.

**Cota inferior**

Para $0<S\le L^{\ast}$, el mayor valor que puede tomar $S$ es $L^{\ast}$.

Sabemos entonces que el peor payoff posible en este caso es:
$$
\begin{aligned}
V(L^{\ast})
&= K-\frac{2r}{2r+\sigma^{2}}\,K \\
&= K\!\left(1-\frac{2r}{2r+\sigma^{2}}\right) \\
\end{aligned}
$$

Como $\sigma^{2}>0 \Rightarrow \dfrac{2r}{2r+\sigma^{2}}<1$, en este caso $V>0$.

Para $S\ge L^{\ast}$:
$$
V(S)=\frac{K\sigma^{2}}{2r+\sigma^{2}}
\left(\frac{2r+\sigma^{2}}{2r}\,\frac{S}{K}\right)^{-\frac{2r}{\sigma^{2}}}.
$$

Esta expresión es $>0$ ya que $K>0$, $S>0$, $\sigma^{2}>0$, $r>0$.





```python
import numpy as np
import matplotlib.pyplot as plt

def solucion_analitica(Smax=150, K=50, r=0.05, sigma=0.2, N=1000):
    h = Smax / N
    S = np.linspace(0, Smax, N+1)

    L_star = (2*r / (2*r + sigma**2)) * K

    V = np.zeros_like(S)

    for i in range(len(S)):
        if S[i] <= L_star:
            V[i] = K - S[i]
        else:
            V[i] = (K*sigma**2/(2*r+sigma**2))*(((2*r +sigma**2)/(2*r))*(S[i]/K))**(-2*r/sigma**2)

    plt.figure(figsize=(8,5))
    plt.plot(S, V, label="Solución Analítica")
    plt.axvline(L_star, color="red", linestyle="--", label=f"L* = {L_star:.2f}")
    plt.plot(S, np.maximum(K-S, 0), "--", label="Payoff")
    plt.xlabel("S (Precio del subyacente)")
    plt.ylabel("Valor de la opción")
    plt.title("Voluación con Solución Analítica")
    plt.legend()
    plt.grid(True)
    plt.show()

    return V, L_star

V, L_star = solucion_analitica()
```

# Ejercicio 2 - Parte A - Método de Shooting

```python
import numpy as np
import matplotlib.pyplot as plt

def shooting(k, smax, sigma, r, tol=1e-8, N=4000):


    # Despejamos V¨ de la ecuación y convertimos la ODE en un sistema de grado 1
    # Esto encuentra las derivadas para el rk4
    def rhs_s(s, y):
        V, Vp = y
        Vpp = (2/(sigma**2 * s**2)) * (r*V - r*s*Vp)
        return np.array([Vp, Vpp])

    def rk4_step(f, t, y, h):
        # Computamos la fórmula de runge kutta para cada step en la integración
        k1 = f(t, y)
        k2 = f(t + h/2.0, y + (h/2.0)*k1)
        k3 = f(t + h/2.0, y + (h/2.0)*k2)
        k4 = f(t + h, y + h*k3)
        return y + (h/6.0)*(k1 + 2*k2 + 2*k3 + k4)

    def F(L, historial = False):
        """Integra desde s=L hasta smax y devuelve V(smax)."""
        s0 = L
        y  = np.array([k - L, -1.0])
        h  = (smax - s0)/N
        s  = s0
        hist = []
        for _ in range(N):
            y = rk4_step(rhs_s, s, y, h)
            s += h
            hist.append((s,float(y[0])))
        return ([y[0], hist]) if historial == True else y[0]


    # Buscamos L* con rootfinder
    a, b = 0.01*k, k
    Fa, Fb   = F(a), F(b)

    if Fa*Fb > 0:
        # Busco un intervalo donde cambie el signo, lo que implica que habrá una raíz en el medio
        grid = np.linspace(0.1*k, 0.99*k, 50)

        for a, b in zip(grid[:-1], grid[1:]):
            fa, fb = F(a), F(b)
            if fa*fb <= 0:
                a, b, Fa, Fb = a, b, fa, fb
                break

    # corremos bisección, vamos achicando el intervalo
    for _ in range(300):
        mid = (a + b)/2
        Fmid  = F(mid)
        if abs(b - a) < tol:
            L_star = mid
            break
        if Fa*Fmid <= 0:
            b, Fb = mid, Fmid
        else:
            a, Fa = mid, Fmid
    else:
        L_star = 0.5*(a + b)

    # devolvemos el residuo en smax y el historial del rk4 para el l encontrado
    residuo = F(L_star)
    hist = F(L_star, True)[1]


    return L_star, residuo, hist

K = 50
sigma = 0.2
r = 0.05
Smax = 150
L_guess = 60

L_star, V_Smax, hist = shooting(K, Smax, sigma, r, tol=1e-8, N=4000)
L_teorico = 2*r*K / (2*r+sigma**2)
print("L* ≈", L_star, "   V(Smax) ≈", V_Smax, 'L teórico≈', L_teorico)

S_cont, V_cont = map(np.array, zip(*hist))   # Separamos s y v(s)

# zona de ejercicio
S_ex = np.linspace(0, L_star, 200)
V_ex = K - S_ex

# payoff para comparar
S_all  = np.concatenate([S_ex, S_cont])
payoff = np.maximum(K - S_all, 0)

plt.figure(figsize=(8,5))
plt.plot(S_cont, V_cont, lw=2, label="Solución numérica")
plt.plot(S_ex,  V_ex,  lw=2, label="Ejercicio: V=K-S")
plt.plot(S_all, payoff, "--",  lw=2, label="Payoff")
plt.axvline(L_star, color="r", ls="--", label=f"L* = {L_star:.2f}")
plt.xlabel("S (Precio del subyacente)"); plt.ylabel("Valor de la opción")
plt.title("Valuación por shooting")
plt.grid(alpha=0.3); plt.legend()
plt.xlim(0, max(S_all)); plt.show()
```

# Ejercicio 2 - Parte B - Diferencias Finitas


```python
import matplotlib.pyplot as plt

def Diferencias_Finitas(Smax=150, K=50, r=0.05, sigma=0.2, N=1000, ruido=1e-10, tol=1e-8, max_iter=500, ponderacion=0.1):
    h = Smax / N
    S = np.linspace(0, Smax, N+1)

    # Vector de valores sobre el que vamos a ir iterando y vector de payoffs
    # que luego nos sirve para encontrar el punto L*
    V = np.maximum(K - S, 0.0)
    Payoff = np.maximum(K - S, 0.0)

    for iteracion in range(max_iter):
        Vold = V.copy()

        # Get A y Get b
        A = np.zeros((N+1, N+1))
        A[0, 0] = 1
        A[N, N] = 1

        b = np.zeros(N+1)
        b[0] = K
        b[N] = 0

        for i in range(1, N):
            if K - S[i] > V[i] + ruido:
                # imponemos condición V[i] + ruido = K - S[i]
                A[i, i] = 1
                b[i] = K - S[i]
            else:
                # EDOs
                A[i, i-1] = 0.5 * sigma**2 * S[i]**2 / h**2 - 0.5 * r * S[i] / h
                A[i, i]   = -(sigma**2 * S[i]**2 / h**2 + r)
                A[i, i+1] = 0.5 * sigma**2 * S[i]**2 / h**2 + 0.5 * r * S[i] / h

        V_new = np.linalg.solve(A, b)
        V = (1 - ponderacion) * Vold + ponderacion * V_new
        if np.linalg.norm(V - Vold, np.inf) < tol:
            print(f"Convergió en {iteracion+1} iteraciones")
            break

    # Calculamos L*
    diferencia = [v - p for v, p in zip(V, Payoff)]
    corte = []
    for i in range(len(diferencia)-1):
        if diferencia[i] <= 0 and diferencia[i+1] > 0:
            corte.append(i)
    if len(corte)==0:
        L_estrella_numerico = None
    else:
        L_estrella_numerico = S[corte]

    L_estrella_analitico = (2*r / (2*r + sigma**2)) * K

    plt.figure(figsize=(8,5))
    plt.plot(S, V, label="Valor numérico")
    plt.plot(S, Payoff, "--", label="Payoff")
    if L_estrella_numerico is not None:
        plt.axvline(L_estrella_numerico, linestyle=":", label=f"L* num = {L_estrella_numerico[0]}")
    plt.axvline(L_estrella_analitico, color="red", linestyle="--", label=f"L* ana = {L_estrella_analitico:.2f}")
    plt.xlabel("S (Precio del subyacente)")
    plt.ylabel("Valor de la opción")
    plt.title("Valuación por Diferencias Finitas")
    plt.legend()
    plt.grid(True)
    plt.show()

    return S, V, L_estrella_numerico, L_estrella_analitico

S, V, L_numerico, L_analitico= Diferencias_Finitas()
```

# Comparación y análisis

Como se puede observar en el gráfico que está a continuación, en cuanto a errores, ambos métodos presentan errores pequeños frente a la solución analítica. En este caso particular, el método de diferencias finitas presenta un error relativo significativamente menor que el método de shooting, lo que sugiere una mayor precisión para esta configuración de parámetros. Esto no significa que este sea siempre el caso, dependiendo de la parametrización podría ocurrir que el método de shooting nos de resultados más precisos que el de diferencias.

En cuanto a robustez, para el método de shooting, la misma depende de la elección del intervalo inicial de busqueda de la raíz y de la estabilidad del método de integración, en este caso Runge Kutta. Por otro lado, en cuanto al método de diferencias finitas, su robustez depende de la discretizacion de la malla, y el orden usado, a mayor orden, más precisión debido a un menor truncamiento de la función.

Por ultimo, en cuanto a la sensibilidad, ambos métodos son sensibles al tamaño  y las tolerancias. En el método de shooting, un intervalo inicial inadecuado o un N muy pequeño pueden llevar a no encontrar la raíz o a una solución imprecisa. En el de diferencias finitas, aparte de esto, esta el problema de la elección del factor de ponderación y la de Smax.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1) Solución Analítica ---
def solucion_analitica(Smax=150, K=50, r=0.05, sigma=0.2, N=1000):
    h = Smax / N
    S = np.linspace(0, Smax, N+1)

    L_star = (2*r / (2*r + sigma**2)) * K

    V = np.zeros_like(S)
    for i in range(len(S)):
        if S[i] <= L_star:
            V[i] = K - S[i]
        else:
            V[i] = (K*sigma**2/(2*r+sigma**2))*(((2*r +sigma**2)/(2*r))*(S[i]/K))**(-2*r/sigma**2)
    return S, V, L_star


# --- 2) Método Shooting ---
def shooting(K, smax, sigma, r, tol=1e-8, N=4000):
    def rhs_s(s, y):
        V, Vp = y
        Vpp = (2/(sigma**2 * s**2)) * (r*V - r*s*Vp)
        return np.array([Vp, Vpp])

    def rk4_step(f, t, y, h):
        k1 = f(t, y)
        k2 = f(t + h/2.0, y + (h/2.0)*k1)
        k3 = f(t + h/2.0, y + (h/2.0)*k2)
        k4 = f(t + h, y + h*k3)
        return y + (h/6.0)*(k1 + 2*k2 + 2*k3 + k4)

    def F(L, historial = False):
        s0 = L
        y  = np.array([K - L, -1.0])
        h  = (smax - s0)/N
        s  = s0
        hist = []
        for _ in range(N):
            y = rk4_step(rhs_s, s, y, h)
            s += h
            hist.append((s,float(y[0])))
        return ([y[0], hist]) if historial else y[0]

    a, b = 0.01*K, K
    Fa, Fb = F(a), F(b)

    if Fa*Fb > 0:
        grid = np.linspace(0.1*K, 0.99*K, 50)
        for a, b in zip(grid[:-1], grid[1:]):
            fa, fb = F(a), F(b)
            if fa*fb <= 0:
                Fa, Fb = fa, fb
                break

    for _ in range(300):
        mid = (a+b)/2
        Fmid = F(mid)
        if abs(b-a)<tol:
            L_star = mid
            break
        if Fa*Fmid<=0:
            b, Fb = mid, Fmid
        else:
            a, Fa = mid, Fmid
    else:
        L_star = 0.5*(a+b)

    hist = F(L_star, True)[1]
    S_cont, V_cont = map(np.array, zip(*hist))
    return S_cont, V_cont, L_star


# --- 3) Diferencias Finitas ---
def Diferencias_Finitas(Smax=150, K=50, r=0.05, sigma=0.2, N=1000, ruido=1e-10, tol=1e-8, max_iter=500, ponderacion=0.1):
    h = Smax / N
    S = np.linspace(0, Smax, N+1)

    V = np.maximum(K - S, 0.0)
    Payoff = np.maximum(K - S, 0.0)

    for _ in range(max_iter):
        Vold = V.copy()
        A = np.zeros((N+1, N+1))
        A[0,0], A[N,N] = 1, 1
        b = np.zeros(N+1)
        b[0], b[N] = K, 0
        for i in range(1,N):
            if K - S[i] > V[i] + ruido:
                A[i,i], b[i] = 1, K - S[i]
            else:
                A[i,i-1] = 0.5*sigma**2*S[i]**2/h**2 - 0.5*r*S[i]/h
                A[i,i]   = -(sigma**2*S[i]**2/h**2 + r)
                A[i,i+1] = 0.5*sigma**2*S[i]**2/h**2 + 0.5*r*S[i]/h
        V_new = np.linalg.solve(A, b)
        V = (1-ponderacion)*Vold + ponderacion*V_new
        if np.linalg.norm(V-Vold, np.inf)<tol:
            break

    diferencia = [v-p for v,p in zip(V,Payoff)]
    corte = [i for i in range(len(diferencia)-1) if diferencia[i]<=0 and diferencia[i+1]>0]
    L_estrella_numerico = None if len(corte)==0 else S[corte]
    return S, V, L_estrella_numerico


# --- Comparación ---
K, r, sigma, Smax = 50, 0.05, 0.2, 150

# Analítica
S_a, V_a, L_a = solucion_analitica(Smax, K, r, sigma)
# Shooting
S_s, V_s, L_s = shooting(K, Smax, sigma, r)
# Diferencias finitas
S_f, V_f, L_f = Diferencias_Finitas(Smax, K, r, sigma)

# Interpolamos numéricas para comparar con analítica
V_s_interp = np.interp(S_a, S_s, V_s)
V_f_interp = np.interp(S_a, S_f, V_f)

# Errores relativos
err_shoot = np.linalg.norm(V_s_interp - V_a, np.inf) / np.linalg.norm(V_a, np.inf)
err_DF    = np.linalg.norm(V_f_interp - V_a, np.inf) / np.linalg.norm(V_a, np.inf)

print(f"Error relativo método shooting: {err_shoot:.2e}")
print(f"Error relativo diferencias finitas: {err_DF:.2e}")

# --- Gráfico conjunto ---
plt.figure(figsize=(9,6))
plt.plot(S_a, V_a, label="Analítica", lw=2)
plt.plot(S_s, V_s, "--", label="Shooting", lw=2)
plt.plot(S_f, V_f, ":", label="Dif. Finitas", lw=2)
plt.axvline(L_a, color="r", ls="--", label=f"L* ana = {L_a:.2f}")
plt.axvline(L_s, color="g", ls=":", label=f"L* shoot = {L_s:.2f}")
if L_f is not None:
    plt.axvline(L_f[0], color="b", ls="--", label=f"L* dif = {L_f[0]:.2f}")
plt.xlabel("S (Precio del subyacente)")
plt.ylabel("Valor de la opción")
plt.title("Comparación de métodos: Analítica vs Shooting vs Dif. Finitas")
plt.legend()
plt.grid(True)
plt.show()

```
